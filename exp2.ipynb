```python
samplerate,source = read('data/src/sample1.wav') #45 seconds long 
plt.plot(source)
```

```python
source = source.T[0]
trans = fft(source)
plt.plot(abs(trans[:len(trans)//2]-1),'r')
print(len(trans)/samplerate)
```

```python
slice1 = source[37*samplerate:38*samplerate]
slice2 = source[30*samplerate:31*samplerate]
write('./data/res/audio4.wav',samplerate,slice1.astype(source.dtype))
write('./data/res/audio5.wav',samplerate,slice2.astype(source.dtype))
trans1 = fft(slice1)
trans2 = fft(slice2)

plt.plot(abs(trans1[:5000]),'r', alpha =0.5)
plt.plot(abs(trans2[:5000]),'g', alpha = 0.5)

mix = trans1+trans2

write('./data/res/audio6.wav',samplerate,ifft(mix).astype(source.dtype))
```
